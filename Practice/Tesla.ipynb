### Simple Moving Average
The Simple Moving Average (SMA) is primarily used in financial markets, economics, supply chain management,
quality control, and meteorology because it is simple to calculate,
helps identify and confirm trends, reduces noise, and is versatile across various applications.

```python
# rolling(10): This applies a rolling window of size 10 to the 'High' column.
# A rolling window is a fixed-size window that moves over the data, performing a calculation at each position
data['High'].rolling(10).mean()
```

```python
data['High'].rolling(10).mean().head(20)
```

```python
data['Open:30 days rolling'] = data['Open'].rolling(30).mean()
```

```python
data.head(31)
```

```python
data[['Open','Open:30 days rolling']].plot(xlim=['2021-01-01','2024-01-01'],figsize = (12,5))
```

```python
data['Open: 10 days rolling'] = data['Open'].rolling(window = 10, min_periods=1).mean()
data['Open: 30 days rolling'] = data['Open'].rolling(window = 30, min_periods=1).mean()
data['Open: 50 days rolling'] = data['Open'].rolling(window = 50, min_periods=1).mean()
```

```python
data[['Open','Open: 10 days rolling','Open: 30 days rolling','Open: 50 days rolling']].plot(xlim = ['2022-01-01' ,'2024-07-01'],figsize = (12,6))
```

### Disadvantages of Simple Moving Average (SMA):

1. Lagging Indicator: Slow to react to recent price changes.
2. Sensitivity to Outliers: Affected by sudden, short-term spikes or drops.
3. Equal Weighting: Assigns equal importance to all data points, which may not be ideal.
4. Fixed Window: Arbitrary choice of window size may not adapt to different conditions.
5. No Predictive Power: Only smooths past data without predicting future trends.

### Expanding
#### Cummulative Moving Avearge
Expanding in Time Series refers to a technique used to calculate expanding statistics, 
where the window size increases with each data point, encompassing all previous observations. 
Unlike fixed-window methods like the Simple Moving Average (SMA),
expanding methods take into account all data points up to the current point in time.

```python
data['Open'].expanding().mean().plot(figsize = (12,6))
```

### Disadvantages of Expanding in Time Series
1. Lagging Nature: Similar to SMA, expanding calculations can lag, especially when changes occur rapidly.
2. Computationally Intensive: As the number of data points increases, the computational load also increases, 
   which can be a drawback for very large datasets.

### Exponential Moving Avg
Exponential Moving Average (EMA) is a type of moving average that gives more weight to recent data points, making it more sensitive to recent price movements. It's commonly used in technical analysis to smooth out price fluctuations and identify trends.

EMA = (Close * alpha) + (Previous EMA * (1 - alpha))


```python
# alpha = 0.1 means that the EWM function will give more weight to recent data points
# adjust=False means that the EWM function will not adjust the weights.

data['EMA_0.1'] = data['Open'].ewm(alpha = 0.1, adjust=False).mean()
```

```python
data[['Open','EMA_0.1']].plot(xlim = ['2023-01-01' ,'2024-07-01'],figsize = (12,6))
```

```python
data['EMA_0.3'] = data['Open'].ewm(alpha = 0.3, adjust=False).mean()
```

```python
data[['Open','EMA_0.1','EMA_0.3']].plot(xlim = ['2023-01-01' ,'2024-07-01'],figsize = (12,6))
```

```python
data['EMA_5days'] = data['Open'].ewm(span=5).mean()
```

```python
data[['Open','EMA_0.1','EMA_5days']].plot(xlim = ['2023-01-01' ,'2024-07-01'],figsize = (12,6))
```
